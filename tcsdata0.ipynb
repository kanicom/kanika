```python
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tcs=yf.download('TCS.NS',
      start='2020-01-01',
      end='2025-02-25',
      progress=False)
```

```python
tcs
```

```python
tcs.info()
```

```python
tcs['Close'].plot(figsize=(12,8))
```

```python
tcs['Volume'].plot(figsize=(12,8))
```

```python
tcs[['Close','Volume']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
tcs.isnull()
```

```python
tcs.isnull().sum()
```

```python
tcs['simple_rtn']=tcs.Close.pct_change()
tcs['simple_rtn']
```

```python
tcs['vsimple_rtn']=tcs.Volume.pct_change()
tcs['vsimple_rtn']
```

```python
tcs
```

```python
tcs[['simple_rtn','vsimple_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
tcs['log_rtn']=np.log(tcs.Close/tcs.Close.shift(1))
tcs
```

```python
tcs['vlog_rtn']=np.log(tcs.Volume/tcs.Volume.shift(1))
tcs
```

```python
tcs[['log_rtn','vlog_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
tcs[['simple_rtn','log_rtn']].plot(subplots=True,style='b',figsize=(12,8))
```

```python
WINDOW_SIZE = 12
tcs['rolling_mean'] = tcs.Close.rolling(window=WINDOW_SIZE).mean()
tcs['rolling_std'] = tcs.Close.rolling(window=WINDOW_SIZE).std()
tcs
```

```python
WINDOW_SIZE = 10
tcs['rolling_mean'] = tcs.Close.rolling(window=WINDOW_SIZE).mean()
tcs['rolling_std'] = tcs.Close.rolling(window=WINDOW_SIZE).std()
tcs
```

```python
WINDOW_SIZE = 10
tcs['vrolling_mean'] = tcs.Volume.rolling(window=WINDOW_SIZE).mean()
tcs['vrolling_std'] = tcs.Volume.rolling(window=WINDOW_SIZE).std()
tcs
```

```python
tcs[['rolling_mean','vrolling_mean']].plot(subplots=True,style='b',figsize=(12,8))
```
